```python
import librosa
import os

def extract_chroma_stft(audio_file):
   
    if not os.path.isfile(audio_file):
        print(f"Error: '{audio_file}' is not a valid audio file.")
        return None

 
    try:
        y, sr = librosa.load(audio_file)
    except Exception as e:
        print(f"Error loading audio file: {e}")
        return None
   
    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    
    return chroma_stft


audio_file = '/kaggle/input/audio-path/biden-to-linus.wav'  
chroma_stft_features = extract_chroma_stft(audio_file)
if chroma_stft_features is not None:
    print(chroma_stft_features)
```

```python
import librosa

def extract_rms(audio_file):
    
    y, sr = librosa.load(audio_file)
    
   
    rms = librosa.feature.rms(y=y)
    
    return rms

audio_file = '/kaggle/input/audio-path/biden-to-linus.wav'  
if os.path.exists(audio_file):
    rms_features = extract_rms(audio_file)
    print(rms_features)
else:
    print("Audio file not found at the specified path.")
```

```python
import librosa

def extract_mfcc1(audio_file):
    
    y, sr = librosa.load(audio_file)
    
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=1)
    
    return mfcc[0]  


audio_file = '/kaggle/input/audio-path/biden-to-linus.wav'  
if os.path.exists(audio_file):
    mfcc1_feature = extract_mfcc1(audio_file)
    print(mfcc1_feature)
else:
    print("Audio file not found at the specified path.")
```
